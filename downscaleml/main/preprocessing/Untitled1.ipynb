```python
import numpy as np

# Assuming arr is your 4-dimensional numpy array
arr = np.random.random((10, 20, 30, 40))  # Adjust the shape according to your array
arr.shape
# Get the total number of elements in the array

```

```python
total_elements = np.prod(arr.shape)
total_elements
```

```python

# Reshape the array to a 2-dimensional array
arr_reshaped = arr.reshape((total_elements // arr.shape[-1], arr.shape[-1]))

# Now, arr_reshaped is a 2-dimensional array
print(arr_reshaped.shape)
```

```python
import numpy as np

# Assuming arr_reshaped is your 2-dimensional array
# For demonstration, let's create a random 2D array
arr_reshaped = np.random.random((6000, 40))  # Adjust the shape according to your reshaped array

# Assuming original_shape is the shape of the original ndim=4 array
original_shape = (10, 20, 30, 40)  # Adjust the shape according to your original array

# Reshape the array back to the original ndim=4 shape
arr_original_shape = arr_reshaped.reshape(original_shape)

# Now, arr_original_shape is a 4-dimensional array
print(arr_original_shape.shape)
```
